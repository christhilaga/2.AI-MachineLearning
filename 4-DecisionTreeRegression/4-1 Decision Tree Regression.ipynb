```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
dataset
```

```python
#Change categorical data to numerical and drop dummies value
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
#Get Dataset columns to assign the input and output values
dataset.columns
```

```python
#Assign input and output variable 
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend',
       'State_Florida', 'State_New York']]
dependent=dataset[['Profit']]
independent
```

```python
dependent
```

```python
#split Train and Test data in dataset
#x=input,y=output
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
y_test
```

```python
# Preprocessing Model 
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
x_test
```

```python
#Model Creation - Decision Tree
#Hyper tuning Parameters
from sklearn.tree import DecisionTreeRegressor
#regressor=DecisionTreeRegressor()
regressor=DecisionTreeRegressor(criterion='friedman_mse', splitter='random')
#regressor=DecisionTreeRegressor(criterion='poisson')
#regressor=DecisionTreeRegressor(criterion='absolute_error', splitter='best')
regressor=regressor.fit(x_train,y_train)
```

```python
# Check decision tree grap for our Dataset
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
# Evaluate our model
y_pred=regressor.predict(x_test)
y_pred
```

```python
#Validate out model
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```
