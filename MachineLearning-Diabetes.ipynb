```python
# Train and Test Fit # Package - sklearn for Train and test,preprocess/Class - model_selection/function - train_test_split
from sklearn.model_selection import train_test_split # its use to split dataset as train(70%)and Tese(30%)/ Data = input and Label =Output
```

```python
x=df.iloc[:,df.columns!='Outcome']# Data/iloc-take particular columns in Dataset/":"- Start to end/df.colums!="Outcome-No need output(Label)"
y=df.iloc[:,df.columns=='Outcome']# Label
```

```python
print(x)
print(y)
```

```python
xtrain,xtest,ytrain,ytest =train_test_split(x,y,test_size=0.2)# 20% take for test and 80% take for train(from dataset)
```

```python
xtrain.head()# for validate
ytrain.head()
```

```python
#Algorithm
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
```

```python
model.fit(xtrain,ytrain.values.ravel())# to train the Algorithm
```

```python
predict_out=model.predict(xtest)#to test the Algorithm
```

```python
print(predict_out)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accscore=accuracy_score(predict_out,ytest)
```

```python
print("The Accuracy Score for RF:-",accscore)
```
