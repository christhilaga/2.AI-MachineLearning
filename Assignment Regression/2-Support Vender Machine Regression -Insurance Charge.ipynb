```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
#Change CategoricalData(Nominal)-Use One Hot Encoading/and remove dummies Values
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
# Get columns
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male',
       'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
#split Train and Test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
x_train,x_test,y_train,y_test
```

```python
#Preprocessing
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
x_test
```

```python
#Create Model
from sklearn.svm import SVR
regressor=SVR(C=10000, cache_size=200, coef0=0.0, degree=3, epsilon=0.1, gamma='auto_deprecated', kernel='linear', max_iter=-1, shrinking=True)
#regressor=SVR(C=10000, cache_size=200, coef0=0.0, degree=3, epsilon=0.1, gamma='auto_deprecated',kernel='rbf', max_iter=-1, shrinking=True, tol=0.001, verbose=False)
#regressor=SVR(C=1000, cache_size=200, coef0=0.0, degree=3, epsilon=0.1,gamma='auto_deprecated', kernel='poly', max_iter=-1, shrinking=True,tol=0.001, verbose=False)
#regressor=SVR(C=1000, cache_size=200, coef0=0.0, degree=3, epsilon=0.1, gamma='auto_deprecated', kernel='sigmoid',max_iter=-1, shrinking=True, tol=0.001, verbose=False)   
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_pred,y_test)
r_score
```
